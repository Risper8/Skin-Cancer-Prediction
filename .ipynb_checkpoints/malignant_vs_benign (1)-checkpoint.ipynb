```python
## train encoder
encoder = {}
for idx, name in enumerate(train_directory):
  encoder[name]=idx

## test encoder
encoder = {}
for idx, name in enumerate(test_directory):
  encoder[name]=idx
```

```python
def process_image(path):
    im = Image.open(path)
    im = im.resize((112, 112))
    rgb_im = im.convert('RGB')
    return np.asarray(rgb_im)
```

```python
import pandas as pd
train_data = pd.DataFrame()
train_data['label'] = labels_train
train_data['path'] = images_train
# data['image_pixel'] = data['path'].map(lambda x: np.asarray(Image.open(x).resize((56,56))))
train_data['image_pixel'] = train_data['path'].map(process_image)
train_data['encode_labels'] = train_data['label'].map(encoder)
```

```python
def prepare_for_train(X_train, Y_train):
    # Prepare data for training and testing the model
    train_datagen = ImageDataGenerator(rescale = 1./255,
                                  rotation_range = 10,
                                  zoom_range=20,
                                  width_shift_range = 0.2,
                                  height_shift_range = 0.2,
                                  shear_range = 0.2,
                                  horizontal_flip = True,
                                  vertical_flip = True,
                                  fill_mode = 'nearest'
                                      )
    train_datagen.fit(X_train)
    return X_train, Y_train
```

```python
def prepare_for_test(X_test, Y_test):
    test_datagen = ImageDataGenerator(rescale = 1./255)
    test_datagen.fit(X_test)
    return X_test, Y_test
```

## modelling

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.applications import EfficientNetB0

def create_model(num_classes, freeze_base_model=True):
    base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(112, 112, 3))
    base_model.trainable = not freeze_base_model
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(128, activation='relu')(x)
    outputs = Dense(num_classes, activation='sigmoid')(x)
    model = Model(inputs=base_model.input, outputs=outputs)
    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4, beta_1=0.9, beta_2=0.999, amsgrad=False, epsilon=1e-08)

    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

    print(model.summary())

    return model
```

```python
X_train, Y_train = prepare_for_train(X_train, Y_train)
```

```python
X_test,Y_test = prepare_for_test(X_test, Y_test)
```

```python
Y_train_flat = Y_train.ravel()
Y_train_flat
```

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np
classes = np.unique(Y_train_flat)

class_weights = compute_class_weight(class_weight="balanced", classes=classes, y=Y_train_flat)
class_weights= dict(zip(classes, class_weights))

class_weights
```

```python
def train_model(model, X_train, Y_train, EPOCHS=20):
    early_stop = EarlyStopping(monitor='val_loss', patience=10, verbose=1,
                           mode='auto', restore_best_weights=True)

    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=5,
                              verbose=1, mode='auto')

    history = model.fit(X_train,
                        Y_train,
                        validation_split=0.2,
                        batch_size = 10,
                        epochs = EPOCHS,
                        class_weight=class_weights,
                        callbacks = [reduce_lr, early_stop])
    return history
```

```python
from sklearn.metrics import confusion_matrix
def test_model(model, X_test, Y_test):
    model_acc = model.evaluate(X_test, Y_test, verbose=0)[1]
    print("Test Accuracy: {:.3f}%".format(model_acc * 100))
    y_true = np.array(Y_test)
    y_pred = model.predict(X_test)
    npred = []
    for i in y_pred:
      if i > 0.5:
        npred.append(1)
      else:
        npred.append(0)
    # y_pred = np.array(list(map(lambda x: np.argmax(x), y_pred)))
    clr = classification_report(y_true, npred, target_names= train_data['label'].unique())
    conf_matrix = confusion_matrix(y_true, npred)
    print(clr)
    target_names= train_data['label'].unique()
    # Plot confusion matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
num_classes = 1
model = create_model(num_classes)
```

```python
model_history = train_model(model, X_train, Y_train, 50)
```

```python
test_model(model, X_test, Y_test)
```
